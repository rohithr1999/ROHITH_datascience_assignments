## <font color='red'> K-Means clustering</fontsize>

```python
df2=df.copy()
df2.drop(['ID#'],axis=1,inplace=True)
df2.head()
```

```python
from sklearn.cluster import KMeans
```

```python
# hyperparameter tunning to get appropriate number of clusters
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df2)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss,marker='o')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
# appling KMeans using 2 clustering
kmeans_2=KMeans(n_clusters=2)
kmeans_2.fit(df2)
y_kmeans_2=kmeans_2.fit_predict(df2)

plt.scatter(df2['Balance'], df2['Bonus_miles'], c=kmeans_2.labels_)
plt.show()
```

```python
# converting numpy array into pandas series object
md_2=pd.Series(y_kmeans_2)
df2['clust']=md_2 # creating a  new column and assigning it to new column
df2.head()
```

```python
# To check the cor_relation
df2.corr()
```

```python
print(f'The kmeans_2.cluster_centers are : {kmeans_2.cluster_centers_}')
print(f'The kmeans_2.labels: {kmeans_2.labels_}')
print(f'The kmeans_2.inertia_: {kmeans_2.inertia_}')
```

```python
df2.groupby(df2['clust']).mean()
```

```python
from sklearn.metrics import silhouette_score,davies_bouldin_score
print("Original Dataset Clustering Results:")
silhouette_avg = silhouette_score(df2, y_kmeans_2)
print(f"Silhouette Score: {silhouette_avg}")
davies_bouldin_index = davies_bouldin_score(df2, y_kmeans_2)
print(f"Davies-Bouldin Index: {davies_bouldin_index}")
```

## <font color='red'>DBSCAN</font>

```python
dfd=df.copy()
dfd.drop(['ID#','Award?'],axis=1,inplace=True)
dfd.head()
```

```python
# preprocessing the data using standardscaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
dfd=scaler.fit_transform(dfd)
```

```python
# calculate the pariwise distances
from scipy.spatial.distance import pdist,squareform
distances=pdist(dfd,metric='euclidean') # Pairwise distances
sorted_distances = np.sort(distances)       # Sort distances for visualization

#visualize the Plot distance distribution
plt.figure(figsize=(8, 5))
plt.plot(sorted_distances)
plt.xlabel('sorted pairwise distances')
plt.ylabel('Distances')
plt.title('Pairwise distance distribution')
plt.grid()
plt.show()
```

```python
## Apply DBSCAN
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=1.9,min_samples=10)
db.fit(dfd)
y_db=db.fit_predict(dfd)
y_db

# Assuming 'Balance' was the first column and 'Bonus_miles' the sixth before scaling:
plt.scatter(dfd[:, 0], dfd[:, 5], c=y_db, s=20, cmap=plt.cm.copper_r)
plt.xlabel('Balance')  # Continue using original labels for clarity
plt.ylabel('Bonus_miles')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
#scores
print("Original Dataset Clustering Results:")
silhouette_avg = silhouette_score(dfd, y_db)
print(f"Silhouette Score: {silhouette_avg}")
davies_bouldin_index = davies_bouldin_score(dfd, y_db)
print(f"Davies-Bouldin Index: {davies_bouldin_index}")
```
