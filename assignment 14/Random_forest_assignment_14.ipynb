- From the above box boxplot it is clearly visible the outliers are present in every columns of the dataset which as to be treated properly

```python
# To remove the outliers in the dataset
def remove_outliers_replace_with_median(df):
    for col in df.select_dtypes(include=np.number).columns:
        Q1=df[col].quantile(0.25)
        Q3=df[col].quantile(0.75)
        IQR=Q3-Q1
        lower_bound=Q1-(1.5*IQR)
        upper_bound=Q3+(1.5*IQR)
        median=df[col].median()
        df[col]=df[col].apply(lambda x:median if x<lower_bound or x>upper_bound else x)
        return df
df2=remove_outliers_replace_with_median(df)      
df2
```

```python
# split the data into train and test
from sklearn.model_selection import train_test_split
x=df2.drop(columns=['Type'])
y=df2['Type']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

### Implementation of model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
RF_classifier=RandomForestClassifier(n_estimators=100, max_samples=0.7, max_features=0.7,random_state = 42)
RF_classifier.fit(x_train,y_train)
y_pred_train=RF_classifier.predict(x_train)
y_pred_test=RF_classifier.predict(x_test)
print(f'The accuracy score of train: {np.round(accuracy_score(y_train,y_pred_train),2)}')
print(f'The accuracy score of test: {np.round(accuracy_score(y_test,y_pred_test),2)}')
```

```python
# cross validation 
training_acc=[]
testing_acc=[]
for i in range(1,100):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = i)       # split the data set 
    RF_classifier = RandomForestClassifier(n_estimators=100, max_samples=0.7, max_features=0.7,random_state = 42)
    RF_classifier.fit(x_train,y_train)
    y_predict_train = RF_classifier.predict(x_train)              # predict the train data
    y_predict_test = RF_classifier.predict(x_test)                # predict the est data
    training_acc.append(accuracy_score(y_train,y_predict_train))
    testing_acc.append(accuracy_score(y_test,y_predict_test))
d1 = np.mean(pd.DataFrame(training_acc))
d2 = np.mean(pd.DataFrame(testing_acc))
print('cross validation RF_classifier train accuracy :',np.round(d1,2))
print('cross validation RF_classifier test accuracy :',np.round(d2,2))
```

## Evaluation of performance on testing data

```python
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score
# make predictions
y_pred=RF_classifier.predict(x_test)
y_pred
```

```python
#Evaluation of metrics
print(f'The accuracy:{np.round(accuracy_score(y_test,y_pred),3)}')
print(f'The precision:{np.round(precision_score(y_test,y_pred,average='weighted'),3)}')
print(f'The recall:{np.round(recall_score(y_test,y_pred,average='weighted'),3)}')
print(f'The f1_score:{np.round(f1_score(y_test,y_pred,average='weighted'),3)}')
```

## Additional notes

### 1. Explain Bagging and Boosting methods. How is it different from each other.

**Bagging (Bootstrap Aggregating):** 
Bagging is an ensemble learning technique that aims to improve model accuracy and reduce variance by training multiple models on different random subsets of the data and averaging their predictions.

**Key Benefits:**

- Reduces overfitting (variance reduction).

- Works well with high-dimensional data.

- Handles noisy data efficiently.

**Boosting:**
Boosting is an ensemble technique that improves model accuracy by sequentially training weak models, each focusing on correcting the errors of the previous one.

**Key Benefits:**

- Improves weak learners by correcting mistakes iteratively.

- Achieves higher accuracy compared to bagging in many cases.

- Works well with structured/tabular data.
