```python
# select the numerical values
df_numbers = data.select_dtypes(include='number')
```

```python
# creat a histogram to identify the skewness
plt.figure(figsize = (20,17))
for i,column in enumerate(df_numbers.columns):  # all columns
    plt.subplot(5,4,i+1)
    skew_ness = skew(df_numbers[column].dropna())
    #plt.title(f'{column}\nSkewness: {skew_ness:.2f}')

    sns.histplot(df_numbers[column],kde=True)
    plt.title(f'Histogram of {column} \nskewness value {skew_ness:.2f}')

plt.tight_layout()
plt.show()
```

```python
# creat a boxplot to identify the outliers
plt.figure(figsize = (20,17))
for i,column in enumerate(df_numbers.columns):  # all columns
    plt.subplot(5,4,i+1)
    sns.boxplot(df_numbers[column],vert=True)   # from a boxplot
    plt.title(f'Boxplot of {column} ')

plt.tight_layout()
plt.show()
```

```python
# Remove the outliers and replace with meadian

def remove_outliers_replace_with_median(df_numbers):
    for column in df_numbers.select_dtypes(include=[np.number]).columns:
        Q1 = df_numbers[column].quantile(0.25)
        Q3 = df_numbers[column].quantile(0.75)
        IQR = Q3-Q1
        lower_bound = Q1 - (1.5 * IQR)
        upper_bound = Q3 + (1.5 * IQR)

        median = df_numbers[column].median()
        df_numbers[column] = df_numbers[column].apply(lambda x: median if x < lower_bound or x > upper_bound else x)
    return df_numbers
clean_data = remove_outliers_replace_with_median(df_numbers)
clean_data
```

```python
object = data['letter']
object.shape
```

```python
from sklearn.preprocessing import LabelEncoder
scaler = LabelEncoder()
object = scaler.fit_transform(object)
object = pd.DataFrame(object)
object.columns = ['letter']
object.head()
```

```python
neurol_data = pd.concat([object,clean_data],axis = 1).dropna()
neurol_data
```

```python
# pairplot
sns.pairplot(neurol_data)
plt.show()
```

```python
y = neurol_data['letter']
x = neurol_data.drop('letter',axis = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler
mm_scaler = MinMaxScaler()
x_scaled = mm_scaler.fit_transform(x)

x = pd.DataFrame(x_scaled, columns=x.columns)

x
```

Model Implementation

```python
# 	Divide the dataset into training and test sets.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import tensorflow as tf
model = tf.keras.Sequential([
    tf.keras.layers.Dense(6, activation='relu', input_shape=(x_train.shape[1],)),   # input layer
    tf.keras.layers.Dense(9, activation='relu'),                          # hidden layer
    tf.keras.layers.Dense(1),                                    # output layer
])
```

```python
model.compile(loss = 'mse',optimizer = 'adam',metrics = ['mae'])
```

```python
history = model.fit(x_train,y_train,epochs = 300,validation_data = (x_test,y_test))
train_loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
import matplotlib.pyplot as plt
train_mae = history.history['mae']
test_mae = history.history['val_mae']
plt.plot(train_mae, label='Train MAE')
plt.plot(test_mae, label='Test MAE')
plt.xlabel('Epochs')
plt.ylabel('MAE')
plt.title('training and test MAE')
plt.legend()
```

```python
from sklearn.metrics import accuracy_score
model.fit(x_train,y_train)
y_pred = model.predict(x_test)  # predict th test data
y_pred
```

```python
# After making predictions:
y_pred = np.round(y_pred)  # predictions to the nearest integer (class label)

ss = accuracy_score(y_test, y_pred)
print('accuracy score of test data ', np.round(ss, 3))
```

```python
# Evaluate the model
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from sklearn.neural_network import MLPClassifier


# 2 hidden layers, 32 neurons each, tanh activation, custom learning rate
model = Sequential()
model.add(Dense(32, input_dim=x_train.shape[1], activation='tanh'))  # Hidden Layer 1
model.add(Dense(32, activation='tanh'))  # Hidden Layer 2
model.add(Dense(1, activation='sigmoid'))  # Output Layer

optimizer = Adam(learning_rate=0.001)

# Compile the model
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2)

# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {accuracy}")
# Example with Scikit-learn’s MLPClassifier - Tuning Hyperparameter
mlp = MLPClassifier(hidden_layer_sizes=(64, 32, 16),
                    activation='tanh',
                    solver='adam',
                    learning_rate_init=0.001,
                    max_iter=500,
                    random_state=42)

mlp.fit(x_train, y_train)
y_pred = mlp.predict(x_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix,f1_score,recall_score,precision_score
print(confusion_matrix(y_test,y_pred))
```

```python
accuracy = accuracy_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred,average = 'weighted')
recall = recall_score(y_test,y_pred,average = 'weighted')
precision = precision_score(y_test,y_pred,average = 'weighted')
print('accuracy score of test data :',accuracy)
print('f1 score :',f1)
print('recall score :',recall)
print('precision score :',precision)
```

```python
# Plot confusion matrix heatmap
plt.figure(figsize=(15, 9))
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Test Data')
plt.tight_layout()
plt.show()
```
