## <font color='orange'>Feature Engineering</font>

```python
# To remove outliers in the data
def remove_outliers_replace_with_median(df1):
    for column in df1.select_dtypes(include=[np.number]).columns:
        Q1=df1[column].quantile(0.25)
        Q3=df1[column].quantile(0.75)
        IQR=Q1-Q3
        inner_bound=Q1-[1.5*IQR]
        outer_bound=Q3+[1.5*IQR]
        median=df1[column].median()
        df1[column]=df1[column].apply(lambda x:median if x<inner_bound or x>outer_bound else x)
        return df1
df2=remove_outliers_replace_with_median(df1)
df2.head()
```

```python
# split the data in x and y
x=df2.iloc[:,0:16]
y=df2['type']
```

```python
# Data parition
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
# apply kNeigbhors algorthims using machine learning library
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=1)
model.fit(x_train,y_train)
```

```python
# To predict the values for train and test data and also teh accuracy score
from sklearn.metrics import accuracy_score
y_train_pred=model.predict(x_train)
y_test_pred=model.predict(x_test)
train_acc=accuracy_score(y_train,y_train_pred)
test_acc=accuracy_score(y_test,y_test_pred)
print(f'The training accuracy : {np.round(train_acc,3)}')
print(f'The testing accuracy : {np.round(test_acc,3)}')
```

```python
# cross validation along with the value of k
training_accuracy_list=[]
testing_accuracy_list=[]


for k in range(1,18,2):
    training_accuracy=[]
    testing_accuracy=[]


    for i in range(1,100,1):
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
        model = KNeighborsClassifier(n_neighbors=k)
        model.fit(x_train.values,y_train)
        y_train_pred=model.predict(x_train.values)
        y_test_pred=model.predict(x_test.values)
        training_accuracy.append(accuracy_score(y_train,y_train_pred))
        testing_accuracy.append(accuracy_score(y_test,y_test_pred))
    training_accuracy_list.append(np.mean(training_accuracy).round(2))
    testing_accuracy_list.append(np.mean(testing_accuracy).round(2))
    
print("Training Accuracies:", training_accuracy_list)
print("Test Accuracies:", testing_accuracy_list)
        
```

```python
# appropriate distance metric and value for Kfrom sklearn.model_selection import cross_val_score

from sklearn.model_selection import cross_val_score
# Define the range of k values you want to test
k_values=range(1,21)

# Store the mean cross-validated scores
mean_scores=[]

# Try each value of k and calculate cross-validation scores
for k in k_values:
    model=KNeighborsClassifier(n_neighbors=k)
    score=cross_val_score(model,x,y,cv=5,scoring='accuracy')
    mean_scores.append(np.mean(score))

# Print the best k value and the corresponding mean score
best_k = k_values[np.argmax(mean_scores)]
best_score = np.max(mean_scores)
print(f"Best k value: {best_k} with accuracy: {best_score:.2f}")
```

```python
# Evaluate performance using classification report(accuracy, precision, recall, and F1-score)
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=1)
model.fit(x_train,y_train)
y_test_pred=model.predict(x_test)
print("Evaluate the performance using classification report")
print(classification_report(y_test,y_test_pred))
```
