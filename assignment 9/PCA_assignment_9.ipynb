```python
# removing of outliers
def remove_outliers_replace_with_medians(df):
    for column in df.select_dtypes(include=[np.number]).columns:
        Q1=df[column].quantile(0.25)
        Q3=df[column].quantile(0.75)
        IQR=Q3-Q1
        lower_bound=Q1-(1.5*IQR)
        upper_bound=Q3+(1.5*IQR)
        median=df[column].median()
        df[column]=df[column].apply(lambda x:median if x<lower_bound or x>upper_bound else x)
    return df
df1=remove_outliers_replace_with_medians(df)
df1
```

## <font color='blue'> 2.Dimensionality reduction using PCA</font>

```python
df2=df1.copy()
df2.head()
```

```python
# separating the cataagorical column
cat_data=df2['Type']
cat_data
```

```python
# standardize the original dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
variables = ['Alcohol', 'Malic', 'Ash', 'Alcalinity', 'Magnesium', 'Phenols',
       'Flavanoids', 'Nonflavanoids', 'Proanthocyanins', 'Color', 'Hue',
       'Dilution', 'Proline']
df2[variables]=scaler.fit_transform(df2[variables])
df2
```

```python
# only numeric columns
df3=df2.iloc[:,1:]
df3
```

```python
# applying principle component analysis
from sklearn.decomposition import PCA
pca=PCA()
pc_data=pca.fit_transform(df3)
pc_data=pd.DataFrame(pc_data)
pc_data
```

```python
# To check pca varience
pc_data.var()
```

```python
pca.explained_variance_
```

```python
pca_data=pc_data.iloc[:,:3]
pca_data
```

```python
# concatenation of transform of data
final_pca=pd.concat([pca_data,cat_data],axis=1)
final_pca.columns=['PC_1','PC_2','PC_3','Type']
final_pca
```

```python
#4.	Analyze the trade-offs between using PCA and clustering directly on the original dataset.
'''
##Clustering on Original Dataset
Dimensionality:
Better suited for low-dimensional data where the original features are directly meaningful.

Interpretability:
Easier to interpret clusters, as they directly relate to the original features, making the results more understandable.

Correlation Between Features:
May suffer from redundancy due to correlated features, which can distort the clustering process.

Computational Cost:
Higher computational cost, especially for large datasets with many features, as clustering must process all dimensions.

Cluster Quality:
Can capture all dimensions, including those with low variance, which may be important for identifying clusters.



##PCA Before Clustering
Dimensionality:
Effective for high-dimensional data, as it reduces the number of features while retaining most of the variance.

Interpretability:
Harder to interpret clusters, as they are defined in terms of principal components, which are linear combinations of original features.


Correlation Between Features:
Removes multicollinearity by transforming correlated features into uncorrelated principal components.

Computational Cost:
Reduces computational cost by lowering the dimensionality of the data.

Cluster Quality:
May lose clustering-relevant dimensions if those dimensions have low variance and are not preserved in the principal components.
'''
```
