<p style="font-size:20px;font-family:'times new roman';">1. What is lift and why is it important in Association rules?</p>

Lift is a metric used in association rule mining to measure how much more likely two items are to be bought together compared to random chance. It helps determine the strength of an association between items. Lift measures whether items truly have a strong association or if they appear together just by chance. Lift is crucial in association rule mining because it helps identify strong, useful, and non-random relationships between products.

<p style="font-size:20px;font-family:'times new roman';">2. What is support and Confidence. How do you calculate them?</p>

Support is calculated by dividing the number of transactions containing a specific itemset by the total number of transactions, while confidence is calculated by dividing the support of the combined itemset by the support of the antecedent itemset Represents the overall popularity of an itemset within the dataset. 

Formula: Support(X) = (Number of transactions containing X) / (Total number of transactions Represents the likelihood of an item appearing given that another specific item is already present.

Formula: Confidence(X -> Y) = Support(X ∪ Y) / Support(X)
